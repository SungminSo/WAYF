```python
import numpy as np
import tensorflow as tf
from PIL import ImageOps
from PIL import Image
```

```python
def image_to_csv(path, label):
    with open(path, 'rb') as f:
        with Image.open(f) as img:
            arr = np.array(img)[:,:,:3]
            return ",".join(map(str,arr.flat)) + f",{label}"
```

```python
path = "C:/Users/ploff/WAYF/Data/valid_pictures/Albania/170612.png"
image_to_csv(path, "Albania")
```
